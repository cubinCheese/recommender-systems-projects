```python
def readGz(path):
    i = 0
    for l in gzip.open(path, 'rt'):
        if i < 500000:
            yield eval(l)
            i += 1
```

# Read in first 500k data points in the beer rating dataset

```python
data = []

for d in readGz("ratebeer.json.gz"):
    data.append(d)
```

```python
data[0]
```

# Filter for datapoints in the top 5 categories

```python
filtered_data = []

top_5_categories = ['India Pale Ale &#40;IPA&#41;', 'Pale Lager', 'Belgian Strong Ale', 'Imperial Stout', 'Imperial/Double IPA']

for d in data:
    if d['beer/style'] in top_5_categories:
        filtered_data.append(d)
        
filtered_data[0]
```

```python
data = filtered_data
```

```python
# convert data into pandas dataframe
df = pd.DataFrame(data)
df.head()
```

```python
df.shape
```

```python
# filter for the first 100k examples
df = df.iloc[0:100000, ]
df.shape
```

```python
df.head()
```

```python
# data cleaning steps
df['beer/ABV'] = df['beer/ABV'].replace({'-': np.nan})
df['beer/ABV'] = df['beer/ABV'].astype(np.float64)
df['review/appearance'] = df['review/appearance'].apply(lambda x: int(x.split('/')[0]) / int(x.split('/')[1]))
df['review/aroma'] = df['review/aroma'].apply(lambda x: int(x.split('/')[0]) / int(x.split('/')[1]))
df['review/palate'] = df['review/palate'].apply(lambda x: int(x.split('/')[0]) / int(x.split('/')[1]))
df['review/taste'] = df['review/taste'].apply(lambda x: int(x.split('/')[0]) / int(x.split('/')[1]))
df['review/overall'] = df['review/overall'].apply(lambda x: int(x.split('/')[0]) / int(x.split('/')[1]))
```

```python
df.head()
```

```python
# plot scatter matrix of data
pd.plotting.scatter_matrix(df, figsize = (12, 12));
```

```python
def beer_features(review):
    '''create a feature vector based on beer features'''
    
    abv = review['beer/ABV']
    appearance = review['review/appearance']
    aroma = review['review/aroma']
    palate = review['review/palate']
    taste = review['review/taste']
    time = int(review['review/time'])
    
    features = [abv, appearance, aroma, palate, taste, time]
    return features
```

```python
def beer_styles(review):
    '''return beer style to use as label for model training'''
    
    return review['beer/style']
```

```python
df = df.dropna()
df.shape
```

```python
X = list(df.apply(beer_features, axis = 1))
y = list(df.apply(beer_styles, axis = 1))
```

```python
# split data into train and validation sets
Xtrain = X[:9 * len(X) // 10]
ytrain = y[:9 * len(X) // 10]
Xvalid = X[9 * len(X) // 10:]
yvalid = y[9 * len(X) // 10:]
```

```python
X[0]
```

```python
model = ensemble.RandomForestClassifier(n_estimators = 100)
model.fit(Xtrain, ytrain)
predictions = model.predict(Xvalid)
accuracy = sum(predictions == yvalid) / len(predictions == yvalid)
```
